**Strategy Overview**

The UPRO/SPXU duel strategy is deterministic and follows these rules:

1. **Morning Cleanup**: Close any overnight positions at 09:30
2. **PDT Check**: Verify we can trade without violating PDT rules
3. **Duel Entry**: Buy equal $ amounts of UPRO and SPXU at 09:30
4. **Winner Window**: Monitor 5-minute window (09:31-09:35) for 0.5% spread
5. **Close Loser**: Sell losing ETF when winner is declared
6. **Close Winner**: Exit winner at next day open OR same day close (configurable)

No machine learning model is required. All logic is implemented in the backtest.
